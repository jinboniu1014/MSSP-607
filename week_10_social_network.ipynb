```python
import numpy as np
import networkx as nx
from matplotlib import pyplot as plt
```

```python
# Create a graph
BTS_graph = nx.Graph()

# Add BTS members as nodes
bts_members = ["RM", "Jin", "Suga", "J-Hope", "Jimin", "V", "Jungkook"]
BTS_graph.add_nodes_from(bts_members)

# Add relationships between BTS members with weights
relationships = [
    ("RM", "Jin", 4),
    ("RM", "Suga", 3),
    ("RM", "J-Hope", 2),
    ("Jin", "Suga", 1),
    ("Suga", "J-Hope", 3),
    ("J-Hope", "Jimin", 2),
    ("Jimin", "V", 4),
    ("V", "Jungkook", 3),
    ("Jungkook", "Jimin", 1),
    ("V", "J-Hope", 2),
]

# Add weighted edges
for edge in relationships:
    BTS_graph.add_edge(edge[0], edge[1], weight=edge[2])

# Visualize the graph with edge weights
pos = nx.spring_layout(BTS_graph, seed=42)
labels = {member: member for member in BTS_graph.nodes()}

edge_labels = {edge: weight for edge, weight in nx.get_edge_attributes(BTS_graph, "weight").items()}

# Draw nodes, edges, and edge labels
nx.draw(BTS_graph, pos, with_labels=True, labels=labels, node_size=1900, font_size=8, font_color='black')
nx.draw_networkx_edge_labels(BTS_graph, pos, edge_labels=edge_labels)
plt.title("BTS Social Network Graph with Edge Weights")
plt.show()
```
