```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/UPenn/MSSP-607/Modules/Week7/Internet Sales.csv')
df.head()
```

## Use Google Colabs and Python to create a bar chart of yearly sales from the Internet Sales Data Set

### Check the data type of order date

```python
type(df['OrderDate'][1])
```

### Get the year from 'OrderDate' by substring method

```python
def get_short(data):
  return data[:4]

df["Year"] = df["ShipDate"].apply(get_short)

```

```python
yearly = df.groupby("Year")
yearlysale = yearly['SalesAmount'].agg('sum')
yearlysale
```

### Visualize yearly sales amount by bar chart

```python
yearlysale = np.log10(yearlysale)
yearlysale.plot(kind='bar')
plt.savefig('/content/drive/MyDrive/UPenn/MSSP-607/Modules/Week7/myfig.pdf')
```
